```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df.drop(columns=['agent', 'company'], inplace=True)
```

```python
df.describe(include='object')
```

```python
for col in df.describe(include='object').columns:
    print(col)
    print(df[col].unique())
    print('-'*60)
```

```python
df['reservation_status_date'] = pd.to_datetime(df['reservation_status_date'])
```

```python
df.dropna(subset=['country', 'children'], inplace=True)
```

```python
df.isna().sum()
```

```python
df.groupby(['customer_type'])['adr'].mean().plot(kind='bar') #adr stands for Average Daily Rate
```

```python
sns.countplot(data=df, x='distribution_channel', hue='is_canceled')
plt.legend(['Not Cancelled', 'Cancelled'])
```

```python
df.groupby('hotel')['is_canceled'].sum().plot(kind='bar')
```

```python
sns.countplot(data=df, x='customer_type', hue='is_canceled')
plt.legend(['Not Cancelled', 'Cancelled'])
```

```python
df[df['is_canceled'] == 0].groupby(['country'])['adr'].size().sort_values(ascending=False).head(10)
# These are the top 10 countries that from which people book hotels
```

```python
year = df['reservation_status_date'].dt.year
month = df['reservation_status_date'].dt.month
```

```python
plt.figure(figsize=(20, 8))
sns.lineplot(data=df, x='reservation_status_date', y='adr', hue='hotel').set_title('ADR OF RESORT AND CITY HOTEL')
plt.show()
```

```python
df.groupby([year,month])['adr'].mean().sort_values(ascending=False).head(10)
```

```python
df.groupby([month])['adr'].sum()
```

```python
df[df['is_canceled'] == 0].groupby([month])['adr'].mean()
```

```python
# Below we can see that cancelled bookings are mostly because of the high price that has been charged.
```

```python
df[df['is_canceled'] == 1].groupby([month])['adr'].mean()
```

```python
df['market_segment'].value_counts(normalize=True)
```

```python
df.groupby(['market_segment'])['adr'].sum().plot(kind='bar')
```

```python
plt.figure(figsize=(17, 8))
sns.barplot(data=df, x='market_segment', y='adr', hue='is_canceled')
plt.show()
```

```python
plt.figure(figsize=(17, 8))
sns.countplot(data=df, x='market_segment', hue='is_canceled')
plt.legend(['Not Canceled','Canceled'])
plt.show()
```
